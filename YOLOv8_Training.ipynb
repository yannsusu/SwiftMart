```python
from ultralytics import YOLO

model = YOLO("yolov8l.pt")

model.train(
    data="/content/drive/My Drive/Colab/SwiftMart/dataset/data.yaml",
    epochs=30,
    imgsz=640,
    batch=16,
    device="cuda",
    workers=4,
    name="yolov8l_custom"
)
```
